```python
# Kmeans Cluster Creation
# import KMeans from sklearn

from sklearn.cluster import KMeans
```

```python
# Create an instance of KMeans with 2 clusters

kmeans = KMeans(n_clusters=2)
```

```python
# Fit the model to all the data except from the Private label

kmeans.fit(df.drop('Private', axis=1))
```

```python
# Evaluation 
# Create a new column for df called 'Cluster', which is 1 for a private school , and a 0 for public school
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
df.Cluster
```

```python
# Kmeans centers

kmeans.cluster_centers_
```

```python
# Kmean labels

kmeans.labels_
```

```python
# Create confusion Matrix and Classification Report

from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(df['Cluster'], kmeans.labels_))
```

```python
# confusion matrix

confusion_matrix(df['Cluster'], kmeans.labels_)
```
