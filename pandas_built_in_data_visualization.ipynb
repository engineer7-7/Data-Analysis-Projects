```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv('df1.csv', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2.csv', index_col=0)

df2.head()
```

```python
# make a hist into a specific column of the dataframe

df1['A'].hist()
```

```python
df1['A'].plot(
    kind='hist')  # second way to do it . In the kind= I can modify it by typing the kind of plot that I want : bar, hist , etc.
```

```python
# another way
df1['A'].plot.hist()  # faster
```

```python
# plot area
df2.plot.area()
```

```python
# bar
df2.plot.bar()
```

```python
# line plot

df1.plot.line()
```

```python
# scatter

df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# box plot

df2.plot.box()
```

```python
# hexbin

df = pd.DataFrame(np.random.randn(1000, 2), columns=['A', 'B'])
df
```

```python
df.plot.hexbin(x='A', y='B', gridsize=25, cmap='coolwarm')
# the more darker is the color the more values belong there
```

```python
# kde plot

df2.plot.kde()
```

```python
# Or just to draw a single column

df2['b'].plot.kde()
```
