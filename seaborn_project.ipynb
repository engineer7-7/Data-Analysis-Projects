```python
# titanic dataset in this exercise
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
# load titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# its given in the udemy exercise the output plot and i have to figure the code that I have to write in order to have that output.

# fare is given as xlabel and age as ylabel
sns.scatterplot(x='fare', y='age', data=titanic)
```

```python
# second
# only fare is given as xlabel

sns.histplot(titanic['fare'])
```

```python
# third
# xlabel is given as dass
sns.boxplot(x='class', y='age', data=titanic, hue='class')
```

```python
# forth

sns.swarmplot(x='class', y='age', data=titanic, hue='class')  # it suggests to use striplot for better usage.
```

```python
# fifth

sns.countplot(x='sex', data=titanic, hue='sex')
```

```python
# sixth
import numpy as np

correlation_matrix = titanic.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True)
```

```python
# last one 

fg = sns.FacetGrid(data=titanic, col='sex')
fg.map(sns.histplot, 'age')
```
