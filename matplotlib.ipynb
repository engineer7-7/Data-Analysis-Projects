```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# I can use whether %matplotlib inline in the beginning or plt.show at the end of the code to plot the data
```

```python
%matplotlib inline
```

```python
# first example
x = np.linspace(0, 5, 11)
y = x ** 2
# functional method
plt.plot(x, y)

# add xlabel , ylabel names
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')

# add title 
plt.title('Square')

```

```python
# Create multi-plots
plt.subplot(1, 2,
            1)  # first number is the number of the rows and the second number is the number of the columns and the third number referrs to the number of plot
plt.plot(x, y, 'r')  # 'r' indicates the color of the plot line. In this case the color is red.
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')  # 'b' indicates the color blue.
```

```python
# Object oriented method

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
plt.title('Square')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.2])
```

```python
# Plot appearance
x = np.linspace(0, 5, 11)
y = x ** 2
plt.plot(x, y, color='purple', linewidth=3,
         linestyle='--')  # linewidth shows us how fat is the line.It can also be written as lw . linestyle shows us the type of the line . --- or ... etc . Linestyle can also be written as ls

```

```python
# Markers
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 2
plt.plot(x, y, 'r', marker='o', markersize=5,
         markerfacecolor='yellow')  # markers indicates what kind of symbols will be in our plotting data . I can also use other marker symbols such as '*' , '+' , etc. I can modify the marker , using markersize= . I can also change the marker color using markerfacecolor='name of the color'

```

```python
# Special plot types. Scatter
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(x, y)

```

```python
# Hist diagramms
from random import sample

data = sample(range(1, 1000), 100)
plt.hist(data, color='red')
```

```python
# Boxplots

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Rectangular boxplot 
plt.boxplot(data , vert=True , patch_artist=True);
```
